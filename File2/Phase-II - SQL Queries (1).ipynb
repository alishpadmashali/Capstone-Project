**Creating a Database Named Shoe Brand**

```python
# Creating  a SQLite database (if it doesn't exist)
conn = sqlite3.connect('ShoeBrand.db')
cursor = conn.cursor()
```

 **Creating 3 tables inside DataBase Shoe_Brand Named Table1 ,Table2 and Table3**

```python
# Create Table1
cursor.execute('''
    CREATE TABLE IF NOT EXISTS Table1 (
        sno INTEGER PRIMARY KEY,
        ShoeName TEXT,
        Category TEXT,
        no_of_colors INTEGER,
        Price REAL,
        Product_code TEXT
    )
''')



# Create Table2
cursor.execute('''
    CREATE TABLE IF NOT EXISTS Table2 (
        sno INTEGER PRIMARY KEY,
        count_of_size REAL,
        Color1 TEXT,
        Color2 TEXT,
        Color3 TEXT,
        Color4 TEXT,
        Color5 TEXT,
        Product_code TEXT
    )
''')

# Create Table3
cursor.execute('''
    CREATE TABLE IF NOT EXISTS Table3 (
        sno INTEGER PRIMARY KEY,
        reviews REAL,
        Size TEXT,
        comfort REAL,
        quality REAL,
        star_rating REAL,
        Product_code TEXT
    )
''')

# Commit the changes 
conn.commit()


print("SQLite database 'Shoe_Brand' and tables 'Table1,' 'Table2,' and 'Table3' created successfully.")
```

**Uploading all the the 3 csv's inside their respected tables**

```python
# Read CSV files into DataFrames
Table1_df = pd.read_csv('Table1.csv')
Table2_df = pd.read_csv('Table2.csv')
Table3_df = pd.read_csv('Table3.csv')

# Upload data to respective tables
Table1_df.to_sql('Table1', conn, if_exists='replace', index=False)
Table2_df.to_sql('Table2', conn, if_exists='replace', index=False)
Table3_df.to_sql('Table3', conn, if_exists='replace', index=False)

# Commit changes 
conn.commit()

print("Data from CSV files uploaded to their respective tables in the 'Shoe_Brand' database.")
```

# 7 Join SQL Queries using all 3 tables

**1.Write a query finds the top-rated men's shoes along with their sizes from "Table1" and "Table3."**

```python
query = ''' SELECT
    T1.ShoeName,
    T3.Size,
    T3.star_rating
FROM Table1 AS T1 JOIN Table3 AS T3 ON T1.Product_code = T3.Product_code
WHERE T1.Category = 'Men' ORDER BY T3.star_rating DESC LIMIT 25;'''
    
# Execute the SQL query and store the result in a DataFrame

result_df = pd.read_sql_query(query, conn)


# Print the DataFrame
result_df
```

**2.Write a query calculates the average comfort rating for each category from "Table1" and "Table3."**

```python
query = '''
SELECT T1.Category,AVG(T3.comfort) AS average_comfort_rating
FROM Table1 AS T1
JOIN Table3 AS T3 ON T1.Product_code = T3.Product_code
GROUP BY T1.Category;
'''

# Execute the SQL query and store the result in a DataFrame
result_df = pd.read_sql_query(query, conn)


# Print the DataFrame
result_df
```

**5.Write a query joins Table 1  and Table 2  using the "Style code/Product code" column, allowing you to retrieve shoe information along with product details.**
