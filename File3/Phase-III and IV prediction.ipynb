**Introduction to the Dataset:**

The dataset comprises a collection of 6 brands of shoes, catering to both men and women. The primary objectives are to predict the Price and Star Rating of the shoes. The dataset consists of 1518 rows and 16 columns.

1. **ShoeName (Column 1):**
   - Indicates the name of the shoe model for men or women.

2. **Category (Column 2):**
   - Represents the category to which the shoe belongs.

3. **Number of Colors (Column 3):**
   - Reflects the count of colors available for a particular shoe model.

4. **Price (Column 4):**
   - Represents the price of the shoes in USD ($).

5. **Count of Sizes (Column 5):**
   - Indicates the number of sizes available for a particular shoe model.

6. **Color1 to Color5 (Columns 6-10):**
   - These columns detail the colors available for the shoes.

7. **Product Code (Column 11):**
   - Represents a unique identifier for each pair of shoes.

8. **Reviews (Column 12):**
   - Indicates the number of reviews received for a particular pair of shoes.

9. **Size (Column 13):**
   - Represents the sizes available for the shoes, measured in UK sizes.

10. **Comfort (Column 14):**
    - Indicates the comfort level of the shoes on a scale of 1 to 10.

11. **Quality (Column 15):**
    - Represents the quality level of the shoes on a scale of 1 to 10.

12. **Star Rating (Column 16):**
    - Indicates the star rating received for the shoes on a scale of 1 to 5.

# Train Test split

### Finding best random state

```python
maxaccu=0
maxRS=0

for i in range(0,200):
    X_train,X_test,y_train,y_test= train_test_split(X,y,random_state=i,test_size=.33)
    LR= LinearRegression()
    LR.fit(X_train,y_train)
    pred= LR.predict(X_test)
    acc=r2_score(y_test,pred)
    if acc>maxaccu:
        maxaccu=acc
        maxRS=i
print("Best accuracy is ",maxaccu,"on Random State =",maxRS)
```

```python
X_train,X_test,Y_train,Y_test= train_test_split(X,y,test_size=.33,random_state=maxRS)
```

```python
# Create instances of regression models
LR_model = LinearRegression()
RFR_model = RandomForestRegressor()
GB_model = GradientBoostingRegressor()
ADA_model = AdaBoostRegressor()
Bag_model = BaggingRegressor()
DT_model = DecisionTreeRegressor()
KNR_model = KNeighborsRegressor()
SVR_model = SVR()

# Create a list of regression models
model = [LR_model, RFR_model, GB_model, ADA_model, Bag_model, DT_model, KNR_model,SVR_model ]
```

```python
for reg_model in model:
    reg_model.fit(X_train, Y_train)
 
    test_pred = reg_model.predict(X_test)
    
    #r2_score
    r2_test = r2_score(Y_test, test_pred)

    # Mean Squared Error (MSE)
    mse_test = mean_squared_error(Y_test, test_pred)
    mae_test = mean_absolute_error(Y_test, test_pred)

    print('Model:', reg_model)
    print('R-squared (Test):', r2_test)
    print('Mean Squared Error (MSE) - Test:', mse_test)
    print('Mean Absolute Error (MAE) - Test:', mae_test)
    print('*' * 100)
```

#### RandomForest is giving 64.56% accuracy so i am considering this for hyperparameter tuning.

```python
param_grid = {
    'n_estimators': [50, 100, 150],
    'criterion': ['squared_error','absolute_error','friedman_mse','poisson'],
    'max_depth': [3, 4, 5],
    'min_samples_split': [2, 3, 4],
    'min_samples_leaf': [1, 2, 4],
}
param_grid
```

```python
grid_search=GridSearchCV(estimator=RFR_model,param_grid=param_grid,cv=5,n_jobs=-1,verbose=2)
grid_search.fit(X_train,Y_train)
```

```python
grid_search.best_params_
```

```python
GCV_pred=grid_search.best_estimator_.predict(X_test)
r2_score(Y_test,GCV_pred)
```
